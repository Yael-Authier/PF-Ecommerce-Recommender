# Hallazgos iniciales

La tabla principal para la construcción del sistema de recomendación será `order_products__prior`, ya que contiene el historial completo de productos comprados por los usuarios.

La relación entre `orders` y `prior` permitirá construir la matriz usuario-producto necesaria para entrenar modelos de recomendación personalizados.

Las tablas `products`, `aisles` y `departments` permitirán enriquecer las recomendaciones con información descriptiva y categórica.

```python
# Calcular cantidad de usuarios únicos

orders["user_id"].nunique()
```

```python
# Calcular cantidad de productos únicos

products["product_id"].nunique()
```

```python
# Calcular cantidad de pedidos únicos

orders["order_id"].nunique()
```

```python
# Calcular cantidad total de productos comprados

len(prior)
```

```python
# Calcular tamaño teórico de la matriz usuario-producto

usuarios = orders["user_id"].nunique()
productos = products["product_id"].nunique()

matriz_teorica = usuarios * productos

print(f"Usuarios: {usuarios:,}")
print(f"Productos: {productos:,}")
print(f"Tamaño teórico matriz: {matriz_teorica:,}")
```

```python
# Calcular sparsity del dataset

interacciones = len(prior)

sparsity = 1 - (interacciones / matriz_teorica)

print(f"Sparsity: {sparsity:.4%}")
```

```python
# Cantidad de compras por producto

product_sales = (
    prior.groupby("product_id")
    .size()
    .reset_index(name="purchase_count")
)

product_sales.sort_values(
    by="purchase_count",
    ascending=False
).head(10)
```

```python
# Estadísticas descriptivas de compras por producto

product_sales["purchase_count"].describe()
```

```python
# Analizar comportamiento de los usuarios

# Cantidad de pedidos por usuario

orders_per_user = (
    orders.groupby("user_id")
    .size()
    .reset_index(name="order_count")
)

orders_per_user["order_count"].describe()
```

```python
# Top 10 usuarios con mayor cantidad de pedidos

orders_per_user.sort_values(
    by="order_count",
    ascending=False
).head(10)
```

```python
# Verificar unicidad de claves primarias

print("Usuarios únicos:", orders["user_id"].nunique())
print("Pedidos únicos:", orders["order_id"].nunique())
print("Productos únicos:", products["product_id"].nunique())

print("\n¿order_id es único en orders?")
print(orders["order_id"].is_unique)

print("\n¿product_id es único en products?")
print(products["product_id"].is_unique)
```

```python
# Verificar que los productos de prior existan en products

prior["product_id"].isin(products["product_id"]).all()
```

```python
# Verificar que los pedidos de prior existan en orders

prior["order_id"].isin(orders["order_id"]).all()
```

# Conclusiones de la fase Data Understanding

## Estructura general

El dataset contiene información de usuarios, pedidos, productos y compras históricas, lo que permite modelar interacciones usuario-producto para construir sistemas de recomendación personalizados.

## Volumen de datos

- 206.209 usuarios únicos.
- 49.688 productos.
- 3.421.083 pedidos.
- 32.434.489 interacciones.

## Calidad estructural

Se verificó la unicidad de las claves primarias y la integridad referencial entre tablas.

- order_id es único en la tabla orders.
- product_id es único en la tabla products.
- Todos los productos registrados en compras históricas existen en el catálogo.
- Todos los pedidos registrados en compras históricas existen en la tabla de pedidos.

## Hallazgos preliminares

- La matriz usuario-producto presenta una sparsity aproximada del 99,68%.
- Se observa un comportamiento Long Tail, donde pocos productos concentran gran parte de las compras.
- Los usuarios presentan un comportamiento recurrente de compra, con una mediana de 10 pedidos por usuario.
